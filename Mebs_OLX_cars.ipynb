```python
# Extracting car brands from the 'Car Name' column
df['Car Brand'] = data['Car Name'].str.split(' ', expand=True)[0]

# Counting occurrences of each car brand
popular_brands = df['Car Brand'].value_counts().nlargest(10)  # Selecting the top 10 brands

# Plotting the most popular car brands
plt.figure(figsize=(10, 6))
sns.barplot(x=popular_brands.values, y=popular_brands.index, palette='viridis')
plt.title('Top 10 Most Popular Car Brands')
plt.xlabel('Number of Listings')
plt.ylabel('Car Brand')
plt.show()
```

```python
# Counting occurrences of each fuel type
fuel_distribution = data['Fuel'].value_counts()

# Plotting the fuel type distribution using a pie chart
plt.figure(figsize=(8, 8))
plt.pie(fuel_distribution, labels=fuel_distribution.index, autopct='%1.1f%%', startangle=140, colors=['skyblue', 'lightgreen', 'salmon'])
plt.title('Fuel Type Distribution')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
# Clean 'Price' column and convert to numeric
df['Price'] = df['Price'].str.replace(',', '').astype(float)

# Extracting the year from the 'Year' column
df['Year'] = pd.to_datetime(df['Year'], format='%Y').dt.year

# Calculating the average price for each year
average_price_year = df.groupby('Year')['Price'].mean()

# Plotting the year-wise trend of average prices
plt.figure(figsize=(10, 6))
sns.lineplot(x=average_price_year.index, y=average_price_year.values, marker='o', color='skyblue')
plt.title('Average Car Price Trend by Year')
plt.xlabel('Year')
plt.ylabel('Average Price (PKR)')
plt.grid(True)
plt.show()
```

```python
# Counting the number of listings for each city
listings_count_city = data['Registration city'].value_counts().sort_values(ascending=False)

# Plotting the location analysis
plt.figure(figsize=(12, 8))
listings_count_city.plot(kind='bar', color='lightblue')
plt.title('Number of Car Listings by City')
plt.xlabel('City')
plt.ylabel('Number of Listings')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Counting the number of listings for each transmission type
transmission_count = df['Transmission'].value_counts()

# Plotting the transmission preferences
plt.figure(figsize=(8, 8))
plt.pie(transmission_count, labels=transmission_count.index, autopct='%1.1f%%', startangle=140, colors=['lightblue', 'lightgreen'])
plt.title('Transmission Preferences')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
# Define price ranges
price_ranges = ['Low', 'Medium', 'High']
cut_bins = [0, 500000, 1000000, df['Price'].max()]

# Categorize prices into ranges
df['Price Range'] = pd.cut(df['Price'], bins=cut_bins, labels=price_ranges)

# Plotting the transmission preferences by price range
plt.figure(figsize=(12, 8))
sns.countplot(x='Price Range', hue='Transmission', data=data, palette='coolwarm')
plt.title('Transmission Preferences by Price Range')
plt.xlabel('Price Range')
plt.ylabel('Number of Listings')
plt.legend(title='Transmission')
plt.show()
```
